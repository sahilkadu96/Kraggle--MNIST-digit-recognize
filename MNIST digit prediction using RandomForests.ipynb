```python
import numpy as np
import pandas as pd

from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
df_train = pd.read_csv(r'G:\Coding ninjas\Kaggle projects & submissions\digit-recognizer\train.csv')
df_train
```

```python
df_train.isnull().sum()
```

```python
df_train['label'].value_counts()
```

```python
Xtrain = df_train.drop('label', axis = 1)
Ytrain = df_train['label']
Xtrain.shape
```

```python
df_test = pd.read_csv(r'G:\Coding ninjas\Kaggle projects & submissions\digit-recognizer\test.csv')
df_test
```

```python
Xtest = df_test
Xtest.shape
```

# Model building

```python
lr = LogisticRegression(penalty = 'l2', max_iter = 1000000)
grid = {'solver':['lbfgs', 'newton-cg', 'sag']}

bestLR = GridSearchCV(lr, grid, cv=5)
bestLR.fit(Xtrain, Ytrain)
bestLR.best_estimator_
```

```python
rf = RandomForestClassifier()
grid2 = {'n_estimators':[10, 50, 100],
         'max_depth': [5, 6, 7]
        }

bestRF = GridSearchCV(rf, grid2, cv=5)
bestRF.fit(Xtrain, Ytrain)
bestRF.best_estimator_
```

```python
bestRF.score(Xtrain, Ytrain)
```

```python
Ypred = bestRF.predict(Xtest)
Ypred
```

```python
df_pred = pd.DataFrame({'ImageId':Xtest.index, 'Label':Ypred})
df_pred
```

```python
df_pred['ImageId'] = df_pred['ImageId'].apply(lambda x:x+1)
df_pred
```

```python
df_pred.to_csv('MNIST_pred.csv', index = False)
```
